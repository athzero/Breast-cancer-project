```python
#separe the dataset to features and target
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
cls_lr = LogisticRegression(random_state=0)
```

```python
cls_lr.fit(x_train, y_train)
```

```python
y_pred = cls_lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, precision_score, recall_score
```

```python
#check the performance of classification model
acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
recal = recall_score(y_test, y_pred)
```

```python
results = pd.DataFrame([['Logistic Regression', acc, f1, prec, recal]], columns = ['Model', 'Accuracy', 'F1_score', 'Precision', 'recall'])
results
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
#Cross Validation
from sklearn.model_selection import cross_val_score
```

```python
accuracies = cross_val_score(estimator=cls_lr, X=x_train, y=y_train, cv=10)
print(accuracies.mean()*100, '%')
```
